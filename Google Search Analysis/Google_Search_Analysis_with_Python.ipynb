#**Import The Necessary Libraries**

```python
import pandas as pd
import pytrends
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

#**Now Let's Find Countries Where Certain Keywords Are Searched The Most**

```python
trends.build_payload(kw_list=['analytics']) #add your keyword here
data = trends.interest_by_region()
data = data.sort_values(by="analytics", ascending=False) #sort by the keyword searched volume
data = data.head(20) #get the top 10 or 20 countries
print(data)
```

```python
data.reset_index().plot(x="geoName", y="analytics", 
                        figsize=(15,10), kind="bar")
# plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['analytics'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 10))
data['analytics'].plot()
plt.style.use('fivethirtyeight')
plt.title("Total Google Searches for keyword: 'Analytics'", fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

#**Historical Interest Overtime**

```python
keywrd_list = ["python programming"]
data = trends.get_historical_interest(keywrd_list, year_start=2020, month_start=1, day_start=1, hour_start=0, year_end=2020, month_end=3, day_end=1, hour_end=0, cat=0, geo='', gprop='', sleep=0)
data = data.sort_values(by="python programming", ascending=False) #sort by the keyword searched volume
data = data.head(30) #get the top 10 or 20 countries
print(data)
```

```python
data.reset_index().plot(x="date", y="python programming", 
                        figsize=(15,6), kind="bar")
# plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['python programming'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 6))
data['python programming'].plot()
plt.title("Total Google Searches for keyword: 'Python Programming' overtime", fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
